```python
import os
import sys

import numpy as  np
import scipy.sparse as sps
import scipy.integrate as spi
import matplotlib.pyplot as plt

# Because python notebooks dont use environment variables
import pyLIQTR
pyliqtr_path = os.path.join(pyLIQTR.__path__[0], 'clam')
if pyliqtr_path not in sys.path:
    sys.path.append(pyliqtr_path)

import pyLIQTR.clam.multilinear_algebra as mla
from pyLIQTR.clam.linearization_tools import linearization_tools as ltools
#import pyLIQTR.clam.linearization_tools.linearization_tools as ltools
import pyLIQTR.clam.linearization_tools.vlasov_spectral as vs

N_x = 4  # Spatial discretization points
N_v = 4 # Velocity discretization points
L_x = 1.0 # Length of spatial domain

gamma = 1.3 # Expansion parameters
epsilon = 1.0

dt = 0.001;
T = 1.0; # Maximum time and timestep
Nstep = int(np.ceil(T/dt)) # Number of points in spatial domain

coeff_set =  vs.vlasov_spectral_system(N_x,N_v,L_x,epsilon,gamma,sparse=True)
coeff_set[0]

F_1_pairs = list(zip(zip(coeff_set[0].row, coeff_set[0].col),coeff_set[0].data))
F_1_pairs

coeff_set[1]
```

```python
N_lin  =  3                 # linearization order

LL = ltools.Linearizer(coeff_set,sparse=True)

LL.carleman(N_lin)          # Build the linearization
#LL.ivp(u_0)                # Build the IVP data

A_s   =  LL.op_s()          # grab the linear evolution matrix
#iv  =  LL.iv()             # grab the initial value vector
A_s
```

```python
from pyLIQTR.ProblemInstances.MatrixInstance import MatrixInstance
from pyLIQTR.BlockEncodings import VALID_ENCODINGS
from pyLIQTR.BlockEncodings.getEncoding import getEncoding

mat_instance = MatrixInstance(A_s)
mat_encoding = getEncoding(VALID_ENCODINGS.PauliLCU, instance=mat_instance)
```
